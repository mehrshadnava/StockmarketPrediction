## ANALYSING effects of delivery ratio change on stock price 
The goal is to study how the delivery ratio will affect the price and and can we use this to predict an positive out break for the next day

```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read and store data in a df
df = pd.read_csv("delivery_traded_ratio_analysis_data.csv")
```

```python
#check missing values
print(df.isnull().sum())
```

```python
#convert datetime string datetime format
df['BH_TIMESTAMP'] = pd.to_datetime(df["BH_TIMESTAMP"])
df = df.sort_values(by=(['BH_SYMBOL','BH_TIMESTAMP']))
```

#### Ploting for TATAMOTORS

```python
df_TATAMOTORS = df[df['BH_SYMBOL']=="TATAMOTORS"].copy()
```

```python
import matplotlib.pyplot as plt

# Set figure size and layout
plt.figure(figsize=(12, 5))

# First subplot: Delivery Traded Ratio
plt.subplot(1, 2, 1)
df_TATAMOTORS['DEL_TRADED_RATIO'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Delivery Traded Ratio")
plt.xlabel("DTR (%)")
plt.ylabel("Frequency")

# Second subplot: Price Change Daily
plt.subplot(1, 2, 2)
(df_TATAMOTORS['Price_Change_Daily']/1000).hist(bins=30, color='lightcoral', edgecolor='black')
plt.title("Distribution of Daily Price Change")
plt.xlabel("Price Change (k)")
plt.ylabel("Frequency")

# Adjust spacing
plt.tight_layout()
plt.show()
```
